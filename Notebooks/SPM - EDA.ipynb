## Processing...

```python
## Maybe Drop NAs?
spms_only['year'].value_counts() - spms_only.dropna()['year'].value_counts()
```

Total of 14 entries...not a lot of data...might as well drop!

```python
# Drop Em!
spms_only.dropna(inplace=True)
```

```python
# Also rename a column for convenience
spms_only.rename(columns={'ahar_part_1_coc_category':'coc_category'}, inplace=True)
```

```python
# Break Down Data by Year, might need but can also always query..
spm_2015 = spms_only.query("year == 2015").copy()
spm_2016 = spms_only.query("year == 2016").copy()
spm_2017 = spms_only.query("year == 2017").copy()
```

```python
spm_2015.describe()
```

```python
# Split Data by SPM
## Get Dict with SPM Numbers as keys and corresponding dfs as values
df_dict = {}
cols = spms_only.columns
spms = ['spm1', 'spm2', 'spm3', 'spm4', 'spm5', 'spm7']
# Keep the first few columns
spm_0 = cols[:8]
## Iterate over columns
for spm in spms:
    index = list(spm_0)+list(spms_only.columns[[spm in col and 'spm0' not in col for col in cols]])
    df_dict[spm] = spms_only[index].copy()
```

```python
df_dict
```

## Visualizations

```python
## Compare Funding and Coc Category Breakdown for Top 5 States by Total COC Award Amount
```

```python
## top 5 States
territories = ['GU', 'VI', 'MP'] # Account for Territories
top_5 = spms_only.query("state not in {}".format(territories)).groupby('state')['spm0_coc_award_amount'].sum().sort_values(ascending=False).index[:6].tolist()
```

```python
## Bottom 5 States
territories = ['GU', 'VI', 'MP'] # Account for Territories
bottom_5 = spms_only.query("state not in {}".format(territories)).groupby('state')['spm0_coc_award_amount'].sum().sort_values().index[:6].tolist()
```

```python
# Graphs
sns.barplot(x='state', y ='spm0_coc_award_amount', hue='coc_category', data = spms_only.query("state in {}".format(top_5)))
```
